Importing the data and finding the mean, median and mode and the percentile.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("income_dataset.csv", names=["name", "income"], skiprows=[0])
df
```

```python
df.describe()
```

```python
df.income.quantile(0.25, interpolation='midpoint')
```

```python
df.income.quantile(0.75, interpolation='lower')
```

```python
df.income.quantile(0, interpolation='higher')
```

```python
persintaile_99 = df.income.quantile(0.99)
persintaile_99
```

```python
df_on_outlier = df[df.income <= persintaile_99]
df_on_outlier
```

Filling the null values/ Data Cleaning

```python
df_mean = df.income.mean()
df_mean
```
